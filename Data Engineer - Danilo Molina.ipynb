```python
username = DataSet['user']
username
```

```python
#Se obtiene el username que existe dentro de los objetos incertos en el arreglo y se obtiene una lista
# Para ello se crea una funcion extract_fullnames_as_string() que  recibe como imput una lista de objetos 
# La funcion entrega una como output una lista de usuarios
```

```python
def extract_fullnames_as_string(list_of_dictionaries):
    return list(map(lambda e : "{}".format(e['username']),list_of_dictionaries))



lista_usuarios = extract_fullnames_as_string(username)
```

## resultado

```python
# se transforma la lista en dataset para operar de forma mas rapida y se utiliza funcion value_counts() para obtener valores
# se utiliza reset_index() para entregar una salida en display que permita una mejor vizualizacion 
```

# Ejercicio 4

```python
# se importa libreria re para aplicar expresiones regulares
```

```python
import re
```

```python
#Se crea nuevo Dataset a partir de la columna content, se obtienen el contenido de los tweets
```

```python
tweets = DataSet['content']
```

```python
# se hace una iteracion a tweets aplicando una expresion regular que permita obtener los hashtags mas utilizados
# la entrada es el dataset tweets y la salida es una lista con todos los hashtags
```

```python
ListaHashtags = []
for i in range(len(tweets)):
    listaTemporal = re.findall(r"#(\w+)", tweets[i])
    ListaHashtags.extend(listaTemporal)
```

En este ejercicio se evaluo la posibilidad de usar List Comprehension y Lambda functions para obtener la lista de hashtags. El primero se descarto debido a que en los tweets donde hay mas de un hashtags entrega una lista de listas, por ende la salida era una lista donde  tenia hashtags, y listas de hashtags de diferentes tamaños, limpiar eso para posteriormente operarlo y filtrarlo significaba perder el tiempo de computo ganado, el cual segun se pudo observar era muy poco en comparacion con el for normal, por el lado de Lambda functions el problema radicaba en errores de poca memoria, lo que indica que no es tan eficiente como los otros metodos

## resultado

```python
# se transforma la lista en dataset para operar de forma mas rapida y se utiliza funcion value_counts() para obtener valores
```

```python
dsHashtags = pd.DataFrame(ListaHashtags , columns =['hashtags'])
dsHashtags['hashtags'].value_counts().nlargest(10).reset_index()
```

# Ejercicio 5 

```python
#se importa libreria emoji, la cual permite identificar una gran cantidad de codigos de emojis mas utilizados
```

```python
import emoji
```

```python
# se crea funcion para encontrar y obtener codigos de emojis
# la entrada es el contenido de los tweets del dataset, el cual fue calculado extraido en el ejercicio 4
# sa salida es una lista con los emojis del tweet
```

```python
def extract_emojis(s):
    return ''.join(c for c in s if c in emoji.UNICODE_EMOJI['en'])
```

```python
#se aplica funcion anterior y se guardan emojis en una lista
# la lista de salida entrega una lista unica con todos los emojis, esta lista es llamada ListaEmoji
```

```python
ListaEmoji = []
for i in range(len(tweets)):
    listaTemporal = extract_emojis(tweets[i])
    ListaEmoji.extend(listaTemporal)
```

## resultado

```python
# se transforma la lista en dataset para operar de forma mas rapida y se utiliza funcion value_counts() para obtener valores
# se muestran los 10 emojis mas utilizados
```

```python
df = pd.DataFrame(ListaEmoji , columns =['Names'])
df['Names'].value_counts().nlargest(10).reset_index()
```
